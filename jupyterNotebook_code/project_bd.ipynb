### Validating the LAW_CAT_CD field

```python
ds_full1.distinct('LAW_CAT_CD')
```

```python
def validate_Law(data,column_name):
    # Check if the offence column just consists of following mentioned offences
    victim_list = ["F", "M", "I", "V"]
    data = data[data[column_name].isin(victim_list)]
    return data
```

```python
data_df = validate_Law(data_df,'LAW_CAT_CD')
#column8[column8['VIC_AGE_GROUP'] == ""]
data_df
```
