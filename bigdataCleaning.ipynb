```python
def check_comp_num(data, columnName):
    # Checking the complaint_number column. Check if all of them are integers 
    comp_numList = list(data[columnName])
    incorrectCompNumList = [s for s in comp_numList if not str(s).isdigit()]
    if len(incorrectCompNumList) > 0:
        for i in incorrectCompNumList:
            data = data[data.CMPLNT_NUM != i]
    else:
        return data
```

```python
column1 = check_comp_num(dataSet, 'CMPLNT_NUM')
```

## Validating Second Column

```python
def validate_date(df,column_name):
    start_date = "01/01/2006"
    end_date = "12/31/2019"

    after_start_date = df[column_name] >= start_date
    before_end_date = df[column_name] <= end_date
    between_two_dates = after_start_date & before_end_date
    filtered_dates = df.loc[between_two_dates]
    return filtered_dates
```

```python
column2 = validate_date(column1,'CMPLNT_FR_DT')
column2
```

## Validating Crime Indicator Column

```python
# Validate the crime indicator column
def validate_crime_indicator(data,column_name):
    # check if the columns consists of crimes mentioned in the below crime_indicator_list
    crime_indicator_list = ["COMPLETED", "ATTEMPTED", "FAILED", "INTERRUPTED PREMATURELY"]
    data = data[data[column_name].isin(crime_indicator_list)]
    return data
```
